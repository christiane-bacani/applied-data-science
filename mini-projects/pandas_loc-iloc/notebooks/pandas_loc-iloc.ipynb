## **Mini-Project #2: Using Pandas Library specifically pandas.DataFrame.loc and pandas.DataFrame.iloc for analyzing [Trending Youtube Video Statistics](https://www.kaggle.com/datasets/datasnaek/youtube-new)**


---

Welcome to the second mini-project I built based on practical concept I learned about for using [Pandas](https://pandas.pydata.org/) library, specifically using [pandas.DataFrame.loc](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html) and [pandas.DataFrame.iloc](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html) for analyzing [Trending Youtube Video Statistics](https://www.kaggle.com/datasets/datasnaek/youtube-new).

This jupyter notebook consist of all the concepts I've learned about using the specific method (**pandas.DataFrame.loc** and **pandas.DataFrame.iloc**) for filtering and manipulating rows and columns of a dataset. For this project, we will use the [**Trending Youtube Video Statistics**](https://www.kaggle.com/datasets/datasnaek/youtube-new) from **Kaggle**.

Specifically, it analyzes the trending youtube videos from **United States of America** that is accumulated from several months (and counting) of daily trending YouTube videos using the method that we've learned for filtering rows.
