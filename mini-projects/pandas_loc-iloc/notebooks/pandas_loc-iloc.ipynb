### **1. Loading dataset**

Load the dataset using **Pandas** library of **Python** and setup the environment to disable scientific notations and use `IPython` library to produce clean output of every code cell.

```python
import pandas as pd
from IPython.display import display

pd.set_option("display.float_format", "{:.2f}".format) # Disable scientific notations

USvideos = pd.read_csv("../datasets/USvideos.csv")
```

### **2. Quick Overview**

Get the conscise summary of the dataset's structure and descriptive statistics of the dataset using `.info()` and `.describe()` respectively.

```python
print("Dataset overview:")
display(USvideos.info())
```

```python
display(USvideos[["views", "likes", "dislikes", "comment_count"]].describe())
```

Get the dataset shape using `.shape` attribute.

```python
display(f"The number of rows and columns: {USvideos.shape}")
```

Check what column or variable consist of **missing values** using `.isnull()` and combine by `.sum()` method to get the total number.

```python
columns = USvideos.columns

for column in columns:
    no_of_missing_values = USvideos[column].isnull().sum()

    if no_of_missing_values > 0:
        display(f"Column: `{column}`")
        display(f"Number of missing values: {no_of_missing_values}")
```

Check the number of unique values for all columns with non-numeric values.

```python
columns = [
    "video_id",
    "trending_date",
    "title",
    "channel_title",
    "publish_time",
    "tags",
    "description"
]

for column in columns:
    no_of_unique_values = len(USvideos[column].dropna().unique())
    display(f"Number of unique values of the column `{column}`: {no_of_unique_values}")
```

Check the unique values of some columns.

```python
for column in columns:
    unique_values = USvideos[column].dropna().unique()

    if column == "tags":
        unique_values = unique_values[:3]

    else:
        unique_values = unique_values[:5]
    
    display(f"Some unique values of column `{column}`:")
    display(unique_values)
```

### **3. Data Cleaning**

Now that we have the overview of what's inside the dataset such as the number of rows and columns, the number of missing values for all columns, unique values per column, and statistical description of the dataset. We will now proceed to clean the dataset based on what we've observed earlier on.

```python
# Handle the missing values of the `description` column
USvideos["description"] = USvideos["description"].fillna("No description")

# Standardize the datatype of `trending_date` and `publish_time` columns
USvideos["trending_date"] = pd.to_datetime(USvideos["trending_date"], format="%y.%d.%m")
USvideos["publish_time"] = pd.to_datetime(USvideos["publish_time"], format="%Y-%m-%dT%H:%M:%S.%fZ")

# Remove unnecessary characters to the `tags` column
USvideos["tags"] = USvideos["tags"].str.replace("[none]", "No tags")
USvideos["tags"] = USvideos["tags"].str.replace("\"", "")
USvideos["tags"] = str(USvideos["tags"].str.split("|").str.join(", "))
```
