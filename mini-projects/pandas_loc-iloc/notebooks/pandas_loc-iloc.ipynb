### **1. Loading dataset**

Load the dataset using **Pandas** library of **Python** and setup the environment to disable scientific notations and use `IPython` library to produce clean output of every code cell.

```python
import pandas as pd
from IPython.display import display

# Disable scientific notations from Pandas library methods
pd.set_option("display.float_format", "{:.2f}".format)

USvideos = pd.read_csv("../datasets/USvideos.csv")
```

### **2. Quick Overview**

Get the conscise summary of the dataset's structure and descriptive statistics of the dataset using `info()` and `describe()` method respectively.

```python
print("Dataset overview:")
display(USvideos.info())
```

```python
display(USvideos[["views", "likes", "dislikes", "comment_count"]].describe())
```

Get the dataset shape using `shape` attribute.

```python
display(f"The number of rows and columns: {USvideos.shape}")
```

Check what columns or variables consist of **missing values** using `isnull()` and combine by `sum()` method to get the total number of values.

```python
columns = USvideos.columns

for column in columns:
    no_of_missing_values = USvideos[column].isnull().sum()

    # Check if the column consist of missing values
    if no_of_missing_values > 0:
        # Display the number of missing values of that column
        display(f"Column: `{column}`")
        display(f"Number of missing values: {no_of_missing_values}")
```

Check the number of unique values for all columns with non-numeric values.

```python
columns = [
    "video_id",
    "trending_date",
    "title",
    "channel_title",
    "publish_time",
    "tags",
    "description"
]

for column in columns:
    # Drop the missing values first before checking the number of unique values
    no_of_unique_values = len(USvideos[column].dropna().unique())
    display(f"Number of unique values of the column `{column}`: {no_of_unique_values}")
```

Check the unique values of some columns.

```python
for column in columns:
    unique_values = USvideos[column].dropna().unique()

    # Display only 3 rows for the `tags` column because it consist of long values
    if column == "tags":
        unique_values = unique_values[:3]

    # Display 5 rows to the rest of columns
    else:
        unique_values = unique_values[:5]

    display(f"Some unique values of column `{column}`:")
    display(unique_values)
```

Check the number of unique values for some columns.

### **3. Data Cleaning**

Now that we have the overview of what's inside the dataset such as the number of rows and columns, the number of missing values for all columns, unique values per column, and statistical description of the dataset. We will now proceed to clean the dataset based on what we've observed earlier on.

```python
# Handle the missing values of the `description` column
USvideos["description"] = USvideos["description"].fillna("No description")

# Standardize the datatype of `trending_date` and `publish_time` columns
USvideos["trending_date"] = pd.to_datetime(USvideos["trending_date"], format="%y.%d.%m")
USvideos["publish_time"] = pd.to_datetime(USvideos["publish_time"], format="%Y-%m-%dT%H:%M:%S.%fZ")

# Remove unnecessary characters to the `tags` column
USvideos["tags"] = USvideos["tags"].str.replace("[none]", "No tags")
USvideos["tags"] = USvideos["tags"].str.replace("\"", "")
USvideos["tags"] = str(USvideos["tags"].str.split("|").str.join(", "))

# Remove duplicates of `video_id` column which is the unique identifier of the dataset
USvideos.drop_duplicates(subset=["video_id"], keep="last", inplace=True)
```

### **4. Data Analysis**

After performing **data cleaning**, we can now proceed to analyze and gain insights by performing data filtering using [**pandas.DataFrame.loc**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html) and [**pandas.DataFrame.iloc**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iloc.html) to the dataset.

Get the row with the index label of `40944` from the dataset.

Get all rows starting from the index label `10` to `76` (inclusive) using index label slicing.

Get all the rows with an index label of `40944`, `40945`, `40946`, and `40947`.

Get all the rows with `views` greater than or equal to **1,000,000** views using boolean masking.

```python
high_viewed_USvideos = USvideos.loc[USvideos["views"] >= 1000000]
display(high_viewed_USvideos)
```

Get the **top 10 most viewed US YouTube Videos**.

```python
# Sort the row based from highest to lowest views
highest_viewed_USvideos = high_viewed_USvideos.sort_values(by="views", ascending=False)
top_10_highest_viewed_USvideos = highest_viewed_USvideos[:10] # Extract only 10 rows

# Save the result to a separate csv file using to_csv() method
top_10_highest_viewed_USvideos.to_csv("../outputs/top_10_highest_viewed_USvideos.csv", index=False)
display(top_10_highest_viewed_USvideos)
```

Get all the rows of the dataset that the year from `publish_time` column starts from **2018** and up.

```python
# Extract only the values of `channel_title` column after performing boolean masking from first argument.
display(USvideos.loc[USvideos["publish_time"].dt.year >= 2018, "channel_title"])
```

Get all the rows that the year from `publish_time` are from `2015` to `2018` (inclusive).

```python
# Perform boolean masking to get all the rows with specified year range from `publish_time`
US_videos_2015_and_up = USvideos.loc[USvideos["publish_time"].dt.year >= 2015]
US_videos_2015_to_2018 = US_videos_2015_and_up.loc[US_videos_2015_and_up["publish_time"].dt.year <= 2018, ["title", "views", "likes", "dislikes", "comment_count"]]

# Save the result to a separate csv file using to_csv() method
US_videos_2015_to_2018.to_csv("../outputs/US_videos_2015_to_2018.csv", index=False)
display(US_videos_2015_to_2018)     
```

Get the most dislike US YouTube Videos with more than `1,000,000` views and extract only the values of specified columns: `channel_title`, `title`, `publish_time`, `views`, `dislikes`, `likes`, and `comment_count`.

```python
# Perform boolean masking to get all the rows with more than 1,000,000 views
most_dislike_USvideos = USvideos.loc[USvideos["dislikes"] > 1000000, ["channel_title", "title", "publish_time", "views", "dislikes", "likes", "comment_count"]]

# Save the result to a separate csv file using to_csv() method
most_dislike_USvideos.to_csv("../outputs/most_dislike_USvideos.csv", index=False)
display(most_dislike_USvideos)
```
