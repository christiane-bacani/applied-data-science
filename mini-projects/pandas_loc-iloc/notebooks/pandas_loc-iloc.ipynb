### **1. Loading dataset**

```python
import pandas as pd
pd.set_option("display.float_format", "{:.2f}".format) # Disable scientific notations

USvideos = pd.read_csv("../datasets/USvideos.csv")
```

### **2. Quick Overview**

```python
print("Dataset overview:")
USvideos.info()
```

```python
USvideos[["views", "likes", "dislikes", "comment_count"]].describe()
```

```python
print(f"The number of rows and columns: {USvideos.shape}")
```

```python
columns = USvideos.columns

for column in columns:
    no_of_missing_values = USvideos[column].isnull().sum()

    if no_of_missing_values > 0:
        print(f"Column: `{column}`")
        print(f"Number of missing values: {no_of_missing_values}")
```
