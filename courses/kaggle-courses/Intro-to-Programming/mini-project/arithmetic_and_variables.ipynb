### **Variables**

[**Variables**](https://www.w3schools.com/programming/prog_variables.php) are identifier in a program used to assign a specific value with a specific memory address assign to it. We can assign a wide range of values with different [**datatypes**](https://www.geeksforgeeks.org/dsa/data-types-in-programming/). Variables are essential in programming to provide a location to a value without needing to write the whole value every single-time we want to use them for different processes.

#### **Creating our own variables**

We can create our own variables by assigning the name to the corresponding value using the equal (`=`) sign like this `variable = value`. We can change the `variable` to name of the variable we want and the `value` with the corresponding value we assign to the variable.

There are certain rules when we cerate our own variable, below is the rules we need to comply to create our own variable:

- **Naming Rules**: Variable names can include letters (A-Z), digits (0-9), and special characters like underscores (_). However, the name of the variable must always start with a lowercase letter.

- **No spaces**: Variables cannot contain whitespaces.

- **Keywords Restrictions**: [Reserved keywords](https://www.w3schools.com/python/python_ref_keywords.asp) cannot be use as a variable.

- **Initialization**: Variables should be initialized with a value before using it to avoid unexpected behviors or errors.

Going back to the previous code cell above, we can create our own variable by assigning the value to it using the result of the arithmetic operations we write.

```python
# Creating our own variables

# The variable 'sum' has a value of the result of the sum of 4 + 4
sum = 4 + 4
print(sum)

difference = 4 - 2
print(difference)

product = 8 * 10
print(product)

quotient = 25 / 5
print(quotient)
```

#### **Manipulating our own variables**

After successfully creating our own variable, we can do multiple things to it, and one of them is called **Variable Manipulation**.

**Variable Manipulation** is the process of manipulating the variable in multiple methods such as variable reassignment, augmented assignment operation, multiple assignment, type conversino, string manipulation, and etc. For our mini-project we will perform the variable reassignment for simplistic approach also.

The code cell below, reassign multiple variables that we've created earlier on when we perform **variable creation**.

```python
# Reassign the 'sum' variable by performing addition to the existing value and 4
sum = sum + 4
print(sum)

# Reassign the 'difference' variable from 2 to 3
difference = 3
print(difference)

# Reassign the 'product' variable from 80 to a string 'This is now a string.'
product = 'This is now a string.'
print(product)

# Reassign the 'quotient' variable from 5.0 to the quotient of the operation 10 / 2
quotient = 10 / 2
print(quotient)
```

#### **Using multiple variables**

We can also use multiple variables to do different operations such as **Arithmetic Operations** that we've discussed earlier on, or **Variable Manipulation** just like the code cell above. This is useful for much more larger task that involved multiple variables. This allows for better organization, readability, and maintanability of the code. Additionally, using multiple variables can help in managing data more effectively, such as tracking multiple values or parameters within a single function or block.

```python
# Reassign the 'product' variable first
product = 80

# Using multiple variables by adding them all and assign to the variable 'result'
result = sum + difference + product + quotient
print(result)
```

The code below uses multiple variables to perform arithmetic operations and using parentheses to manipulate the order of the operation. In this way, we know what operation will be executed first.

```python
result = product - ((sum * difference) / quotient)
print(result)
```

In the code above, this is the order of execution:

- `sum * difference`
- `(sum * difference) / quotient`
- `product - ((sum * difference) / quotient)`

This means that the product of this code `sum * difference` will be executed first, then the result of the product will be divided to the `quotient` variable. Lastly, the result of the division will be subtracted to the `product` variable.

### **Comments**

[**Comments**](https://www.w3schools.com/programming/prog_comments.php) in programming is a text in code that is ignored by the [**interpreter**](https://www.techgeekbuzz.com/blog/what-is-interpreter/) or [**compiler**](https://builtin.com/software-engineering-perspectives/compiler). It is helpful because it helps us understand the code and what it does. This is crucial because sometimes we are not writing the code itself and we want to understand the code that we are reading by reading the comment.

It's like a sticky note to an object to provide description and understand what it does and provide documents or purpose to the logic of code. Comments can be used also to temporarily disable code for **debugging**.
