Welcome to the **Intro to Programming** course! This course is for you if you have never written a single line of code, and you are interested in learning data science and machine learning. (*If you do have programming experience and are just new to the Python language, the [Python course](https://www.kaggle.com/learn/python) is a better fit to get started.*)

In this course, you will learn how to use code to get a computer to perform certain tasks for you. **Python** is one of the most popular programming languages for data science, and it's the language you'll learn in this course. Once you complete this course, you'll be fully prepared to progress to the [Python course](https://www.kaggle.com/learn/python), followed by the [Intro to Machine Learning course](https://www.kaggle.com/learn/intro-to-machine-learning).

In this tutorial, you'll see several examples of Python code. You'll get a chance to write your own code in the exercise. (*if you'd like to preview the exercise, feel free to [take a look now](https://www.kaggle.com/code/christianeabacani/exercise-arithmetic-and-variables/edit).*)
